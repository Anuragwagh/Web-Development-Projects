```python
# importing libraries
import pandas as pd
import scipy
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('C:/Users/rocke/Downloads/export (1).csv')
print(df.head())
```

```python
#Drop null values
def dropvalue(df):
    round((df.isnull().sum()/df.shape[0])*100,2)
    df = df.fillna("")
    df = df.dropna(how='all')
    df.duplicated()

#find outlier
def findoutlier(df):
    plt.boxplot(df['rental_price'], vert=False)
    plt.ylabel('Variable')
    plt.xlabel('rental_price')
    plt.title('Box Plot')
    plt.show()
```

```python
def drawbarchart(df):
    plt.bar(df['days_on_market'], df['rental_price'])
    plt.title("Bar Chart")
    plt.show()
    
def drawscatterplot(df):
    plt.scatter(df['days_on_market'], df['rental_price'])
    plt.title("Scatter Plot")
    plt.show()
    
def drawhistogramt(df):
    plt.hist(df['days_on_market'])
    plt.title("Histogram")
```

```python
def operation(no):
    if(no==1):
        dropvalue(df)
    elif(no==2):
        drawbarchart(df)
    elif(no==3):
        drawscatterplot(df)
    elif(no==4):
        drawhistogramt(df)
    elif(no==5):
        drawbarplot(df)
    elif(no==6):
        findoutlier(df)
    elif(no==7):
        df.info()
    elif(no==8):
        print(df.info())
        print(dropvalue(df))
        print(findoutlier(df))
        drawbarchart(df)
        drawscatterplot(df)
        drawhistogramt(df)
    else:
        print("Give input of numbers from 1 to 8")
```
